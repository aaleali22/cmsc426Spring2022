```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
```

```python
def loadCIFAR10(dataFile):
    hf = h5py.File(dataFile, 'r')
    # Show the data stred in the data file
    # Data is stored as test data, test labels, train data and train labels
    print(hf.keys())
    train_data = hf.get('cifar10_train_data')
    test_data = hf.get('cifar10_test_data')
    train_labels = hf.get('cifar10_train_labels')
    test_labels = hf.get('cifar10_test_labels') 
    # Convert data to numpy arrays
    train_data = np.array(train_data)
    test_data = np.array(test_data)
    # Train and test labels are one-hot encoded
    train_labels = np.array(train_labels)
    test_labels = np.array(test_labels)
    hf.close()
    return train_data, train_labels, test_data, test_labels
    
```

```python
# load cifar10 data file cifar10_data.h5
# train_data,train_labels,test_data,test_labels = loadCIFAR10('cifar10_data.h5')
train_data,train_labels,test_data,test_labels = loadCIFAR10('cifar10_data_small.h5')
```

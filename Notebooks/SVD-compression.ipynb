```python
%matplotlib inline
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('lion.jpg').convert('L')
```

```python
plt.imshow(img, cmap='gray')
```

```python
U,S,V = np.linalg.svd(img,full_matrices=False)
```

```python
print(np.shape(img))
```

```python
print(U.shape)
print(S.shape)
print(V.shape)
```

```python
sigmas=np.diag(S)
```

```python
len(sigmas)
```

```python
lower_rank600 = sigmas
```

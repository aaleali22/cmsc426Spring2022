```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
from sklearn.utils import shuffle
import pandas as pd
```

```python
def plotGauss1D(mu,sigma):
    sd =np.sqrt(sigma)
    x = np.arange(mu-3*sd,mu+3*sd,0.02)
    g = (1/(np.sqrt(2*np.pi)*sd))*np.exp((-0.5*(x-mu)**2)/sigma)
    return x,g
```

```python
data, gaussian = plotGauss1D(0.0,1)
len(data)
```

```python
plt.plot(data,gaussian)
plt.xlabel('x')
plt.ylabel(r"$p(x\,|\,\mu,\sigma^2)$")
```

```python
plt.show()
```

```python
import numpy as np
import pylab as plt
```

```python
n = 1000
theta = np.linspace(0,2*np.pi,n)
x,y = 1 * np.cos(theta),1*np.sin(theta)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x,y,linestyle='-')
plt.ylim([-1.5,1.5])
plt.xlim([-1.5,1.5])
plt.grid()

plt.show(block=True)
```

```python
x.shape = (1000,1)
y.shape = (1000,1)
x.shape
```

```python
y.shape
```

```python
data = np.column_stack((x,y))
```

```python
data.shape
```

```python
s = np.array([[0.3,0],[0,1.4]])
```

```python
s
```

```python
scaled = np.dot(data,s) 
```

```python
scaled.shape
```

```python
plt.figure(figsize=(6,6))
plt.plot(scaled[:,0],scaled[:,1],linestyle='-')
plt.ylim([-1.5,1.5])
plt.xlim([-1.5,1.5])
plt.grid()
plt.show(block=True)
```

```python
R  = np.array([[0.5, 0.866],[-0.866,0.5]])
```

```python
rotated = np.dot(data,R)
```

```python
plt.figure(figsize=(6,6))
plt.plot(rotated[:,0],rotated[:,1],linestyle='-')
plt.ylim([-1.5,1.5])
plt.xlim([-1.5,1.5])
plt.grid()
plt.show(block=True)
```

```python
T = np.array([[0.38,-0.88],[1.18, -0.63]])
transformed = np.dot(data,T)
```

```python
plt.figure(figsize=(6,6))
plt.plot(transformed[:,0],transformed[:,1],linestyle='-')
plt.ylim([-1.5,1.5])
plt.xlim([-1.5,1.5])
plt.grid()
plt.show(block=True)
```
